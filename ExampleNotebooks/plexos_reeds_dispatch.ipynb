# Facet Plots

```python
fig, axs = plt.subplots(2,7, figsize=(25,10))

zones = zone_all.columns.get_level_values(0).unique()
num_zones = len(zones)

axs = qp.trim_axs(axs, num_zones)
i = 0
for ax in axs.reshape(-1)[0:num_zones]:
    
    if i < num_zones:
        zone = zones[i]
        ax = qp.plot_monthly_system_dispatch_stack(zone_all[[zone]], ax)
        ax.set_title(f"{zone}")
        ax.get_legend().remove()

        i = i + 1


#fig.suptitle("Monthly Dispatch")
plt.tight_layout()
    
```
